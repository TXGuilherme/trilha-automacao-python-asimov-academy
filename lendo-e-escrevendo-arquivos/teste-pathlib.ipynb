```python
from pathlib import Path
import os
import ipykernel
```

```python
# Define o caminho da pasta home (Pasta base com todos os arquivos do PC)
pasta_home = Path.home()

# Quando o formato já for Path, é possível concatenar strings com / mantendo o formato
pasta_downloads = pasta_home / 'Downloads'
print(type(pasta_downloads))

print(pasta_home, "\n", pasta_downloads)

# .glob('*') busca todos os arquivos dentro do caminho passado
for arquivo in pasta_downloads.glob('*'):
    print(arquivo)
```

```python
# cwd = current work directory -> Mostra a pasta em que o Python está rodando
Path.cwd().is_absolute()
```

```python
# Encontrando Path da pasta atual com os 
print(os.getcwd())
#Retorno é no formato string 
print(type(os.getcwd()), '\n')

# Encontrando Path da pasta atual com Path
print(Path.cwd())
#Retorno é no formato path
print(type(Path.cwd()))
```

```python
pasta_arquivo = Path.cwd()
print(pasta_arquivo)

print(f'Pasta raiz do computador:{pasta_arquivo.anchor}')
print(f'Pasta anterior: {pasta_arquivo.parent}')

print(pasta_arquivo)
print(f'Função que retorna pastas anteriores de acordo com o índice: {pasta_arquivo.parents[0]}')
```

```python
# os.listdir recebe como argumento um path e retorna tudo dentro dela (Pastas, arquivos e etc)

print(os.listdir(pasta_arquivo / 'pasta_de_teste'))

print(os.listdir(pasta_arquivo))
```

```python
# .glob('*') lista tudo dentro de uma pasta, mas sem entrar nas pastas
print(list(pasta_arquivo.glob('*')))

#.glob('**/*') lista tudo em uma pasta, inclusive, arquivos dentro de pastas contidas na pasta do path. 
print(list(pasta_arquivo.glob('**/*')))
```
